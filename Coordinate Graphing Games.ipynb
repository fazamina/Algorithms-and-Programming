```python
import matplotlib.pyplot as plt 

#Masukkan asumsi awal
plt.xlim([-5, 5]) 
plt.ylim([-5, 5]) 
x = [] 
y = [] 

#Deskripsi program
print("Selamat datang di program corat-coret")
print("Ini ada papan corat-coret masih kosong")
plt.show() 

print("Isi dengan titik koordinat ya") 
print("1. Untuk tambahan titik") 
print("2. Untuk lihat hasil coretan") 
print("3. Untuk hapus semua coretan") 
print("4. Untuk menghentikan program") 

#Pilihan
pilih = int(input("Pilih: "))

#Jika memilih nomor 4
if (pilih == 4): 
    print("Terima kasih telah bermain!") 
else: 
    while (pilih != 4): 
        
        #Jika memilih nomor 1
        if (pilih == 1): 
            x1 = input("Masukkan titik-titik x (dengan format [x1,x2,..,xn]): ") 
            y1 = input("Masukkan titik-titik y (dengan format [y1,y2,..,yn]): ") 
            x += eval(x1) 
            x.append(None) 
            y += eval(y1) 
            y.append(None)
        
        #Jika memilih nomor 2
        elif (pilih == 2) : 
            plt.xlim([-5, 5]) 
            plt.ylim([-5, 5]) 
            plt.plot(x, y, data = None) 
            plt.show()
        
        #Jika memilih nomer 3
        elif (pilih == 3): 
            x = [] 
            y = [] 
        pilih = int(input("Pilih : ")) 
    print("Terima kasih telah bermain!") 
```
