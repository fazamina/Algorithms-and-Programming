```python
#Deskripsi program
print('Selamat datang di F.R.I.D.A.Y.')
print('F.R.I.D.A.Y. merupakan games tebak angka dengan memilih sendiri digit angka yang ingin Anda tebak.')
print(' ')
print('Langkah pertama Anda harus memasukkan digit angka yang akan ditebak')
print('Selanjutnya masukkan batas nyawa yang diinginkan')
print(' ')
print('Selamat Bermain!')
print(' ')

#Panggil modul random
import random 

#Asumsi awal
value = 0 
yes = 0 
pos = 0 
coba = 0 
digit = int(input("Masukkan jumlah digit angka yang ingin ditebak: ")) 
nyawa = int(input("Masukkan batas nyawa yang diinginkan: ")) 
rdm = [0 for i in range (digit)] 

#Angka random awal
for i in range (digit) : 
    rdm[i] = random.randint(0, 9) 

#Angka random [i] digit
rdm2 = ["*" for i in range (digit)]
for i in range (digit) : 
    rdm2[i] = str(rdm[i]) 

#Percobaan tebakan
while (value == 0) : 
    coba += 1 
    yes = 0 
    pos = 0 
    print("Masukkan", digit, "digit angka :", end = " ") 
    isi = input("")
    
    #Memeriksa jumlah digit
    if (len(isi) > digit or len(isi) < digit): 
        print("Jumlah digit harus", digit) 
        value = -1 
    #Looping tebakan yang di isi 
    for i in range (len(isi)): 
        if (isi[i] == '0' or isi[i] == '1' or isi[i] == '2' or isi[i] == '3' or isi[i] == '4' or isi[i] == '5' or isi[i] == '6' or isi[i] == '7' or isi[i] == '8' or isi[i] == '9'): 
            continue 
        else: 
            print("Tebakan hanya boleh angka")
            value = -1 
            break
    if (value == 0) : 
        #Memeriksa keberadaan angka
        for i in range (digit) : 
            ada = 0 
            for j in range (digit) : 
                if (isi[i] == rdm2[j]) : 
                    ada = 1 
            if (ada == 1) : 
                yes += 1 

        #Memeriksa posisi angka
        for i in range (digit) : 
            if (isi[i] == rdm2[i]) : 
                pos += 1 
        if (yes == digit and pos == digit) : 
            value = 1 
            break 
        else : 
            print(yes, "angka benar,", pos, "angka di posisi yang benar") 
            nyawa -= 1 
        if (nyawa == 0) : 
            value = 1 
        else : 
            print("Sisa nyawa :", nyawa) 
    if (value == -1) : 
        value = 0 

#Mencetak output
if (value == 1 and nyawa == 0) : 
    print("Maaf nyawa Anda habis") 
    print("Angka yang benar adalah", rdm2) 
elif (value == 1) : 
    print("Selamat tebakan Anda benar!") 
    print("Anda selesai menebak dengan benar dalam", coba, "percobaan") 
```
